**Data loading & cleaning**

```python
#import library

import pandas as pd
```

```python
df = pd.read_csv ('amazon.csv', encoding_errors='ignore')

df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
#drop row missing values

df.dropna(inplace=True)

#verify

df.isnull().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)

df.head()
```

```python
df['actual_price'] = df['actual_price'].str.replace(r'[₹,]', '', regex=True).astype(float)

df.head()
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=True).astype(int)

```

```python
df.head()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
df.head()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#plot style

sns.set_style("whitegrid")
```

**Distribution of Product Categories**

```python
# Top 10 categories

plt.figure(figsize=(12, 6))
category_counts = df['category'].value_counts().head(10)  
sns.barplot(x=category_counts.values, y=category_counts.index, palette="coolwarm")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.title("Top 10 Product Categories on Amazon")
plt.show()
```

**Price vs Discount Trends**

```python
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(x=df['actual_price'], y=df['discounted_price'], hue=df['discount_percentage'], palette="coolwarm", alpha=0.7)
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.title("Actual Price vs. Discounted Price")
plt.legend(title="Discount Percentage")
plt.show()
```

**Most and Least Rated Products**

```python
top_rated = df.sort_values(by="rating", ascending=False).head(10)
least_rated = df.sort_values(by="rating", ascending=True).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_rated['rating'], y=top_rated['product_name'], palette="Blues_r")
plt.xlabel("Rating")
plt.ylabel("Product Name")
plt.title("Top 10 Highest Rated Products")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=least_rated['rating'], y=least_rated['product_name'], palette="Reds_r")
plt.xlabel("Rating")
plt.ylabel("Product Name")
plt.title("Top 10 Lowest Rated Products")
plt.show()
```

**Lets see, Does Higher Discount Lead to Better Ratings?**

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['discount_percentage'], y=df['rating'], alpha=0.5)
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.title("Discount Percentage vs. Product Rating")
plt.show()
```

**Plotting the bar chart**

```python
#attach catagory & rating

category_sales = df.groupby('category')['rating_count'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Top-Selling Categories')
plt.xlabel('Category')
plt.ylabel('Total Rating Count')
plt.xticks(rotation=45)
plt.show()
```

**Plotting price vs discount trends**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='actual_price', y='discounted_price', marker='o')
plt.title('Price vs Discount Trends')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
#for review column

from wordcloud import WordCloud
```

```python
# Assuming the reviews are stored in the 'review_content' column

text = ' '.join(df['review_content'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plotting the word cloud

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#  'rating' column distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=10, kde=True, color='blue')
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
